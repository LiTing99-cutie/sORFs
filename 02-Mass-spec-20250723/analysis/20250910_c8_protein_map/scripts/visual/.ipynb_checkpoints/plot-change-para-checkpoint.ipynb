```r
# 2) 相关性（原尺度）
pear <- cor.test(df_1$mean_relative_iBAQ, df_1$RPF_RPKM, method = "pearson")
spear <- cor.test(df_1$mean_relative_iBAQ, df_1$RPF_RPKM, method = "spearman")
cat(sprintf("[Pearson] r=%.3f, p=%.2e\n", pear$estimate, pear$p.value))
cat(sprintf("[Spearman] rho=%.3f, p=%.2e\n", spear$estimate, spear$p.value))

# 3) 可选：对数尺度（常见于表达/丰度），加极小伪计数避免 log(0)
eps_x <- max(1e-12, quantile(df_1$mean_relative_iBAQ[df_1$mean_relative_iBAQ>0], 0.001, na.rm=TRUE, names=FALSE))
eps_y <- max(1e-12, quantile(df_1$RPF_RPKM[df_1$RPF_RPKM>0], 0.001, na.rm=TRUE, names=FALSE))

p <- ggplot(df_1, aes(x = mean_relative_iBAQ + eps_x, y = RPF_RPKM + eps_y)) +
  geom_point(alpha = 0.4, size = 1) +
  geom_smooth(method = "lm", se = TRUE, linewidth = 0.6) +
  scale_x_log10(labels = scales::label_number(scale_cut = scales::cut_short_scale())) +
  scale_y_log10(labels = scales::label_number(scale_cut = scales::cut_short_scale())) +
  labs(
    x = "mean_relative_iBAQ (log10, +eps)",
    y = "RPF_RPKM (log10, +eps)",
    title = "Protein (mean_relative_iBAQ) vs Ribo-seq (RPF_RPKM)"
  ) +
  theme_bw(base_size = 12)

# 4) 保存图片
# ggsave("scatter_iBAQ_RPF.png", p, width = 6.5, height = 5.2, dpi = 300)
p
```

```r
filter(orfs,!is.na(mean_relative_iBAQ)&!is.na(RPF_RPKM)) -> df_1
suppressPackageStartupMessages({
  library(dplyr)
  library(purrr)
  library(tidyr)
  library(ggplot2)
})

# ---------- 参数与筛选 ----------
df_2 <- df_1 %>%
#   filter(ORF_type != "canonical") %>%
  filter(Run_occurrence >= 1) %>%
  mutate(
    mean_relative_iBAQ = suppressWarnings(as.numeric(mean_relative_iBAQ)),
    RPF_RPKM = suppressWarnings(as.numeric(RPF_RPKM))
  ) %>%
  filter(is.finite(mean_relative_iBAQ), is.finite(RPF_RPKM))

# ---------- 分组相关（原尺度） ----------
calc_corr <- function(d) {
  n <- nrow(d)
  if (n < 3) {
    tibble(n = n,
           pearson_r = NA_real_, pearson_p = NA_real_,
           spearman_rho = NA_real_, spearman_p = NA_real_)
  } else {
    p <- cor.test(d$mean_relative_iBAQ, d$RPF_RPKM, method = "pearson")
    s <- cor.test(d$mean_relative_iBAQ, d$RPF_RPKM, method = "spearman", exact = FALSE)
    tibble(n = n,
           pearson_r = unname(p$estimate), pearson_p = p$p.value,
           spearman_rho = unname(s$estimate), spearman_p = s$p.value)
  }
}

res_by_type <- df_2 %>%
  group_by(ORF_type) %>%
  group_modify(~ calc_corr(.x)) %>%
  arrange(desc(n))

print(res_by_type)

# ---------- 可选：分组相关（log10 尺度） ----------
calc_corr_log <- function(d) {
  # 全局极小伪计数，避免 log(0)
  eps_x <- max(1e-12, quantile(d$mean_relative_iBAQ[d$mean_relative_iBAQ > 0], 0.001, na.rm = TRUE, names = FALSE))
  eps_y <- max(1e-12, quantile(d$RPF_RPKM[d$RPF_RPKM > 0], 0.001, na.rm = TRUE, names = FALSE))
  x <- log10(d$mean_relative_iBAQ + eps_x)
  y <- log10(d$RPF_RPKM + eps_y)
  ok <- is.finite(x) & is.finite(y)
  n <- sum(ok)
  if (n < 3) {
    tibble(n_log = n, pearson_r_log = NA_real_, pearson_p_log = NA_real_,
           spearman_rho_log = NA_real_, spearman_p_log = NA_real_)
  } else {
    p <- cor.test(x[ok], y[ok], method = "pearson")
    s <- cor.test(x[ok], y[ok], method = "spearman", exact = FALSE)
    tibble(n_log = n, pearson_r_log = unname(p$estimate), pearson_p_log = p$p.value,
           spearman_rho_log = unname(s$estimate), spearman_p_log = s$p.value)
  }
}

res_by_type_log <- df_2 %>%
  group_by(ORF_type) %>%
  group_modify(~ calc_corr_log(.x)) %>%
  arrange(desc(n_log))

print(res_by_type_log)

# ---------- 分面散点图（log10 坐标） ----------
eps_x <- max(1e-12, quantile(df_2$mean_relative_iBAQ[df_2$mean_relative_iBAQ > 0], 0.001, na.rm = TRUE, names = FALSE))
eps_y <- max(1e-12, quantile(df_2$RPF_RPKM[df_2$RPF_RPKM > 0], 0.001, na.rm = TRUE, names = FALSE))

p <- ggplot(df_2, aes(x = mean_relative_iBAQ + eps_x, y = RPF_RPKM + eps_y)) +
  geom_point(alpha = 0.35, size = 0.8) +
  geom_smooth(method = "lm", se = TRUE, linewidth = 0.5) +
  scale_x_log10(labels = scales::label_number(scale_cut = scales::cut_short_scale())) +
  scale_y_log10(labels = scales::label_number(scale_cut = scales::cut_short_scale())) +
  facet_wrap(~ ORF_type, scales = "free") +
  labs(
    x = "mean_relative_iBAQ (log10, +eps)",
    y = "RPF_RPKM (log10, +eps)",
    title = "Non-canonical ORFs: Protein vs. Ribo-seq by ORF_type"
  ) +
  theme_bw(base_size = 12)

# ggsave("scatter_by_ORFtype_iBAQ_vs_RPF.png", p, width = 9, height = 6.5, dpi = 300)
p
```
