```r
bar_plot_basic_stack <- function (df, x, y, stat = "identity", fil_col = NULL, label = FALSE) 
    {
        library(ggpubr)
        library(RColorBrewer)
        library(colorspace)
        p <- ggplot(df, aes_string(x = x, y = y, fill = fil_col)) + 
            geom_bar(stat = stat, width = 0.8, position = position_stack()) + 
            scale_y_continuous(expand = c(0, 0)) + theme_3(rotate = F) + 
            labs(fill = fil_col)
        if (label == TRUE) {
            p <- p + geom_text(aes_string(label = y), position = position_stack(), 
                vjust = 0.5, size = 5, color = "black") + coord_cartesian(clip = "off")
        }
        color_n = n_distinct(df[, fil_col])
        if (!is.null(fil_col)) {
            if (color_n <= 12) {
                p <- p +  scale_fill_manual(values = hcl.colors(color_n, 
                    "Set 2"))
            }
            else {
                p <- p + scale_fill_manual(values = hcl.colors(color_n, 
                    "Set 2"))
            }
        }
        return(p)
    }
df %>% count(ORF_type_1,Scodon) %>% bar_plot_basic_stack(.,x="ORF_type_1",y="n",fil_col  = "Scodon")
```
