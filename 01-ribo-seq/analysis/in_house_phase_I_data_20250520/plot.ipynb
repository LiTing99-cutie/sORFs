```python
# 将数据分割成行，并转换为列表
lines = data.strip().split('\n')
values = []
groups = []

# 解析数据
for line in lines:
    freq, group = line.split()
    if int(group) <= 20:  # 只保留分组在 0 到 20 之间的数据
        values.append(int(freq))
        groups.append(int(group))

# 绘制柱状图
plt.figure(figsize=(10, 6))  # 设置图形大小
plt.bar(groups, values, color='skyblue')  # 创建柱状图

# 设置 x 轴的范围
plt.xlim(0, 20)

# 添加标题和标签
plt.title('Frequency Distribution by Group (0-20)')
plt.xlabel('Group')
plt.ylabel('Frequency')

# 显示图形
plt.show()
```
