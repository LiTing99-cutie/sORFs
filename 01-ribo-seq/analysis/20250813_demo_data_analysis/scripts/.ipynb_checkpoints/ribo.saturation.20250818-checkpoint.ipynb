# 老 gtf 不允许overlap计数

```r
#!/usr/bin/env Rscript
suppressPackageStartupMessages({
  library(data.table); library(dplyr); library(ggplot2)
})

outdir <- "../processed/saturation_tx"
reads_meta <- fread(file.path(outdir, "meta/reads.tsv"))

# 读取所有 featureCounts 输出（从第3行开始是数据）
count_files <- list.files(file.path(outdir, "counts"), pattern="^f\\d+\\.\\d{2}\\.txt$", full.names=TRUE)
if (length(count_files) == 0) stop("No count files found.")

read_fc <- function(fp){
  tag <- sub("\\.txt$", "", basename(fp))
  tag <- sub("^(.+)$", "\\1", tag)
  dt <- fread(fp, skip=2)  # 第1-2行是头信息
  # 最后一列是该样本的计数
  counts <- dt[[ncol(dt)]]
  data.frame(tag=tag, detected_ge1 = sum(counts >= 1),
             detected_ge5 = sum(counts >= 5),
             detected_ge10 = sum(counts >= 10),
             stringsAsFactors = FALSE)
}

det_list <- lapply(count_files, read_fc) %>% bind_rows()
res <- reads_meta %>%
  inner_join(det_list, by="tag") %>%
  arrange(reads_all)

# --------- 绘图（以 >=1 read 判定“覆盖”）---------
p <- ggplot(res, aes(x=reads_all, y=detected_ge1)) +
  geom_point() + geom_line() +
  labs(x="Mapped reads (subsampled)", y="Detected transcripts (≥1 read)",
       title="Transcript rarefaction (featureCounts, -s 1, -O, transcript_id)") +
  theme_classic()
ggsave(file.path(outdir, "saturation_tx_ge1.png"), p, width=6, height=4, dpi=200)

# --------- 简单饱和度判定 ----------
# 规则：末段相邻两个点的增量，占当前总检测量 < 1%
if (nrow(res) >= 3) {
  tail_inc <- diff(tail(res$detected_ge1, 3))
  last_total <- tail(res$detected_ge1, 1)
  sat_flag <- (max(tail_inc) / max(1, last_total)) < 0.01
} else {
  tail_inc <- NA; sat_flag <- NA
}

cat("Saturation (≥1 read) reached? ", sat_flag, "\n", sep="")
cat("Last increments: ", paste(tail_inc, collapse=", "), "\n", sep="")
fwrite(res, file.path(outdir, "saturation_tx_summary.tsv"), sep="\t")
```

```r
library(data.table)
library(dplyr)
library(ggplot2)

# 读你的表
df <- fread("../processed/saturation_tx/saturation_tx_summary.tsv")  # 改成你的文件名
# 选一个阈值列，这里先用 detected_ge5（更稳健）
ycol <- "detected_ge5"
df <- df %>% arrange(reads_all) %>%
  rename(reads = reads_all, y = !!sym(ycol))

# ---- 1) 末段增量判断 ----
dy <- diff(tail(df$y, 3))
dx <- diff(tail(df$reads, 3))
slope_last <- tail(dy, 1) / tail(dx, 1)
incr_rel   <- tail(dy, 1) / tail(df$y, 1)
cat(sprintf("Last slope = %.3e (per read)\n", slope_last))
cat(sprintf("Last relative increment = %.4f\n", incr_rel))

# ---- 2) 模型拟合（指数 & 米氏），选 AIC 小的 ----
start_S <- max(df$y)
start_b <- 1 / max(df$reads)
start_K <- median(df$reads)

fit_exp <- try(nls(y ~ Smax * (1 - exp(-b * reads)),
                   data=df, start=list(Smax=start_S, b=start_b)), silent=TRUE)
fit_mm  <- try(nls(y ~ Smax * reads / (Km + reads),
                   data=df, start=list(Smax=start_S, Km=start_K)), silent=TRUE)

sel <- NULL
if (!inherits(fit_exp,"try-error") && !inherits(fit_mm,"try-error")) {
  aic_exp <- AIC(fit_exp); aic_mm <- AIC(fit_mm)
  sel <- if (aic_exp <= aic_mm) list(model="exp", fit=fit_exp) else list(model="mm", fit=fit_mm)
} else if (!inherits(fit_exp,"try-error")) {
  sel <- list(model="exp", fit=fit_exp)
} else if (!inherits(fit_mm,"try-error")) {
  sel <- list(model="mm", fit=fit_mm)
} else stop("两种模型都拟合失败。")

co <- coef(sel$fit); Smax <- unname(co["Smax"])
if (sel$model=="exp") {
  b <- unname(co["b"])
  x95 <- -log(1-0.95)/b
} else {
  Km <- unname(co["Km"])
  x95 <- (0.95/(1-0.95)) * Km
}
cat(sprintf("Model: %s, Smax≈%.0f, Reads for 95%%≈%.0f\n",
            sel$model, Smax, x95))

# ---- 3) 作图 ----
pred_x <- seq(min(df$reads), max(df$reads), length.out=200)
if (sel$model=="exp") {
  pred_y <- Smax * (1 - exp(-b * pred_x))
} else {
  pred_y <- Smax * pred_x / (Km + pred_x)
}
p <- ggplot(df, aes(x=reads, y=y)) +
  geom_point() +
  geom_line(aes(y=pred_y), data=data.frame(reads=pred_x, pred_y=pred_y)) +
  geom_hline(yintercept=0.95*Smax, linetype=2) +
  geom_vline(xintercept=x95, linetype=2) +
  labs(x="Mapped reads", y=paste0("Detected transcripts (", ycol, ")"),
       title=sprintf("%s fit: Smax≈%.0f; 95%%≈%.0f reads",
                     sel$model, Smax, x95)) +
  theme_classic()
p
ggsave(paste0("saturation_", ycol, ".png"), p, width=6, height=4, dpi=200)
```
