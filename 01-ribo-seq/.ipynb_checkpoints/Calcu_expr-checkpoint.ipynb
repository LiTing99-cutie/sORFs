```r
cus_1 <- function(feature_count_path,lib_size,sub_pattern_1,sub_pattern_2){
read.table(feature_count_path,header = T) -> count
colnames(count[,7:ncol(count)]) %>% gsub(pattern = sub_pattern_1,replacement = "") %>% 
gsub(pattern = sub_pattern_2,replacement = "") -> colnames_new
colnames(count) <- c(colnames(count[,1:6]),colnames_new)
rkm <- count[,7:ncol(count)]/count$Length*1000
rkm %>% t() %>% as.data.frame()-> rkm
rkm %>% tibble::rownames_to_column("Sample") -> rkm
merge(rkm,libsize,by="Sample") %>% 
  mutate(rpkm_cons = V1/libsize*10^6,rpkm_alter = V2/libsize*10^6) %>% 
  mutate(V1=NULL,V2=NULL) %>% 
  mutate(PAU=rpkm_alter/rpkm_cons) %>% 
  dplyr::select(Sample,rpkm_alter,rpkm_cons) -> rpkm
return(rpkm)
}
cus_2_red <- function(x,y){
ggplot(data, aes_string(x = x, y = y)) +
  geom_line(color = "gray", size = 1,group=1) + # 折线
  geom_point(color = "black", fill="#d87c77",size = 5,shape=21) +    # 数据点
  labs(
    x = "Human Brain",
    y = "RPKM"
  ) +
  theme_3()
}
cus_2_blue <- function(data,x,y){
ggplot(data, aes_string(x = x, y = y)) +
  geom_line(color = "gray", size = 1,group=1) + # 折线
  geom_point(color = "black", fill="#2f6bac",size = 5,shape=21) +    # 数据点
  labs(
    x = "Organoids",
    y = "RPKM"
  ) +
  theme_3()
}
```

```r
read.table("/home/user/data/lit/project/ZNF271/data/rna-seq/organoid/mapping/uniq/bam_stat/library_size.txt") -> libsize
colnames(libsize) <- c("Sample","libsize")
cus_1("./fc.case.devo.organoid.txt",
     libsize,
     "X.home.user.data.lit.project.ZNF271.data.rna.seq.organoid.mapping.uniq.",
     ".uniq.sorted.bam") -> rpkm
rpkm$Sample %>% sub(pattern = "_[1-9]+",replacement = "",.) -> rpkm$Day
rpkm %>% group_by(Day) %>% summarize(mean_rpkm=mean(rpkm_alter),median_rpkm=median(rpkm_alter),
                                                    mean_rpkm_cons=mean(rpkm_cons),median_rpkm_cons=median(rpkm_cons),) -> data
data
```

```r
cus_2_blue(data,"Day","mean_rpkm_cons")+ labs( y = "RPKM")
cus_2_blue(data,"Day","mean_rpkm")+ labs( y = "The expression of ORF-containing isoforms (RPKM)")
```
