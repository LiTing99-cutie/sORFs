```python
import sys
import pandas as pd
import numpy as np
```

```python
# 输入文件
inputfile1 = "/home/user/data3/lit/project/sORFs/01-ribo-seq/mouse_brain_output_20241011/2015_Science_ChoJ_hippocampus/SRR2163083/output/Ribo-ORFs/PRICE/SRR2163083.orfs.tsv"
inputfile2 = "/home/user/data3/lit/resource/gtf/mouse/mm39/gencode.vM29.annotation.genePred.txt"
```

```python
inputfile_1 = pd.read_csv(inputfile1, sep='\t', header='infer')
inputfile_2 = pd.read_csv(inputfile2, sep='\t', header=None)
```

```python
print(inputfile_1.head)
print(inputfile_2.head)
```

```python
# 校正p值
import statsmodels.stats.multitest as multi
inputfile_1['FDR']=multi.multipletests(inputfile_1['p value'].values, method="fdr_bh")[1]
```

```python
inputfile_1.head()
```

```python
inputfile_1[inputfile_1['FDR'] <= 0.1]
```

```python
# 筛选FDR小于0.05，起始密码子为ATG，类型是非经典的蛋白质
inputfile_1 = inputfile_1.query("FDR <= 0.05 and Codon == 'ATG' and Type != 'CDS'")
print(inputfile_1.shape[0])
```

```python
inputfile_1["Location"]
```

```python
import re
inputfile_1['gstart']=inputfile_1['Location'].apply(lambda x: min(int(num) for num in re.findall(r'[:|-](\d+)',  x)))
inputfile_1['gstop']=inputfile_1['Location'].apply(lambda x: max(int(num) for num in re.findall(r'[:|-](\d+)', x)))
```

```python
print(inputfile_1.head)
```

```python
type(inputfile_1['Location'].apply(lambda x: re.match(r'(\w+)([+-]):', x).group(1)))
inputfile_1['Location'].apply(lambda x: re.match(r'(\w+)([+-]):', x).group(1)).value_counts()
"chr"+inputfile_1['Location'].apply(lambda x: re.match(r'(\w+)([+-]):', x).group(1))
```

```python
inputfile_1['chr']="chr"+inputfile_1['Location'].apply(lambda x: re.match(r'(\w+)([+-]):', x).group(1))
inputfile_1['strand']=inputfile_1['Location'].apply(lambda x: re.match(r'(\w+)([+-]):', x).group(2))
inputfile_1['ens_id']=inputfile_1['Id'].str.split("_", expand=True).iloc[:,0]
```

```python
print(inputfile_1['strand'].value_counts())
print(inputfile_1['chr'].value_counts())
```

```python
new_id=inputfile_1['ens_id']+inputfile_1['strand']+inputfile_1['chr']+":"+inputfile_1['gstart'].astype(str)+"-"+inputfile_1['gstop'].astype(str)
```

```python
inputfile_1['new_id']=new_id
```
